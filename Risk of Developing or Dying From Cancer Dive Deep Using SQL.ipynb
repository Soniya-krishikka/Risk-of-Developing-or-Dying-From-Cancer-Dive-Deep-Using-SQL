```python
## Import Liberaries
import pandas as pd
import sqlite3
```

```python
## Read the file

df = pd.read_csv('C:\\Users\\subal\\OneDrive\\Desktop\\Documents\\Project\\sql\\Risk of Developing or Dying From Cancer\\Risk of Developing or Dying From Cancer.csv')
```

```python
df.head()
```

```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("cancer_data", con =engine)
```

```python
## Test the code by select the complete table

sql='''
SELECT * FROM cancer_data
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
## Query 2: Select distinct cancer types

sql='''
SELECT DISTINCT "Cancer Type" FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
## Query 3: Count of records per cancer type

sql='''
SELECT "Cancer Type", COUNT(*) as count FROM cancer_data GROUP BY "Cancer Type" ORDER BY count DESC
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Count the number of distinct cancer types in the data table
sql='''
SELECT COUNT(DISTINCT "Cancer Type") AS num_cancer_types FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select cancer types where Male Risk Development Percentage is greater than 10%
sql='''
SELECT "Cancer Type" FROM cancer_data
WHERE "Male Risk Development Percentage" > 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select cancer types and their corresponding Female Risk Dying Percentage
sql='''
SELECT "Cancer Type", "Female Risk Dying Percentage" FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select the average Male Risk Development Percentage 
sql='''
SELECT AVG("Male Risk Development Percentage") AS avg_male_risk_dev_pct FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select cancer types with the highest Male Risk Dying Percentage
sql='''
SELECT "Cancer Type", "Male Risk Dying Percentage" FROM cancer_data
ORDER BY "Male Risk Dying Percentage" DESC
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select cancer types where both Male and Female Risk Development Percentages are greater than 10%
sql='''
SELECT "Cancer Type" FROM cancer_data
WHERE "Male Risk Development Percentage" > 10
AND "Female Risk Development Percentage" > 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select cancer types and calculate the difference between Male and Female Risk Development Percentages
sql='''
SELECT "Cancer Type",
("Male Risk Development Percentage" - "Female Risk Development Percentage") AS risk_dev_diff
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select cancer types and calculate the ratio of Male to Female Risk Development Percentages
sql='''
SELECT "Cancer Type",
("Male Risk Development Percentage" / "Female Risk Development Percentage") AS risk_dev_ratio
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the total Male and Female Risk Development Percentages for all cancer types
sql='''
SELECT SUM("Male Risk Development Percentage") AS total_male_risk_dev_pct,
SUM("Female Risk Development Percentage") AS total_female_risk_dev_pct
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select cancer types and their corresponding Male and Female Risk Dying Ratios
sql='''
SELECT "Cancer Type",
"Male Risk Dying Ratio(One in ____ Person)" AS male_risk_dying_ratio,
"Female Risk Dying Ratio(One in ____ Person)" AS female_risk_dying_ratio
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select cancer types where Male Risk Dying Percentage is less than Female Risk Dying Percentage
sql='''
SELECT "Cancer Type" FROM cancer_data
WHERE "Male Risk Dying Percentage" < "Female Risk Dying Percentage"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select cancer types and their Male and Female Risk Development Percentages, ordered by Female Risk Development Percentage in descending order
sql='''
SELECT "Cancer Type", "Male Risk Development Percentage", "Female Risk Development Percentage"
FROM cancer_data
ORDER BY "Female Risk Development Percentage" DESC
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Calculate the weighted average risk development percentage for males and females, weighted by the risk of dying
sql='''
SELECT 
    SUM("Male Risk Development Percentage" * "Male Risk Dying Percentage") / SUM("Male Risk Dying Percentage") AS weighted_avg_male_risk_dev_pct,
    SUM("Female Risk Development Percentage" * "Female Risk Dying Percentage") / SUM("Female Risk Dying Percentage") AS weighted_avg_female_risk_dev_pct
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Rank cancer types based on combined risk (sum of male and female risk development percentages) and return the top 10
sql='''
SELECT "Cancer Type",
("Male Risk Development Percentage" + "Female Risk Development Percentage") AS combined_risk_dev_pct,
RANK() OVER (ORDER BY ("Male Risk Development Percentage" + "Female Risk Development Percentage") DESC) AS risk_rank
FROM cancer_data
ORDER BY risk_rank
LIMIT 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Calculate the percentage difference between male and female risk dying percentages for each cancer type
sql='''
SELECT "Cancer Type",
(ABS("Male Risk Dying Percentage" - "Female Risk Dying Percentage") / (( "Male Risk Dying Percentage" + "Female Risk Dying Percentage") / 2)) * 100 AS pct_diff_risk_dying
FROM cancer_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Identify cancer types where the male risk of dying is more than twice the female risk of dying
sql='''
SELECT "Cancer Type", "Male Risk Dying Percentage", "Female Risk Dying Percentage"
FROM cancer_data
WHERE "Male Risk Dying Percentage" > 2 * "Female Risk Dying Percentage"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Find the cancer type with the maximum difference in risk development ratios between males and females
sql='''
SELECT "Cancer Type",
ABS("Male Risk Development Ratio(One in ____ Person)" - "Female Risk Development Ratio(One in ____ Person)") AS diff_dev_ratio
FROM cancer_data
ORDER BY diff_dev_ratio DESC
LIMIT 1
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```
